# Line plots

```python
import plotly.graph_objects as go
df_stocks = px.data.stocks()
df_stocks.head()
```

```python
px.line(df_stocks, x='date', y='GOOG', labels={'x': 'Date', 'y':'Price'})
```

```python
px.line(df_stocks, x='date', y=['GOOG', 'AAPL'], labels={'x': 'Date', 'y':'Price'}, title='Apple vs Google')
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_stocks.date, y=df_stocks.AAPL, mode='lines', name='Apple'))
fig.add_trace(go.Scatter(x=df_stocks.date, y=df_stocks.AMZN, mode='lines+markers', name='Amazon'))
fig.add_trace(go.Scatter(x=df_stocks.date, y=df_stocks.GOOG, mode='lines+markers', name='Google', line=dict(color='firebrick', width=1, dash='dashdot')))

fig.update_layout(title = 'Stock Price (2018-2020)')
fig.update_layout(xaxis = dict(showline=False, showgrid=False, showticklabels=True), autosize=False)
```

```python
fig.update_layout(xaxis = dict(showline=True, showgrid=True, 
                               showticklabels=True,
                              tickfont=dict(family='Arial', size=12, color='red')))
```
